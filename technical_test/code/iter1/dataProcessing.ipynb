```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/Users/abilfad/Desktop/legit/technical_test/data/Turbine_Data.csv')
data.info()
data['time'] = pd.to_datetime(data['time'])
data.index = data['time']
```

```python
sns.heatmap(data.isna())
```

```python
data.isnull().sum()/118224
```

```python
data_day = data.resample('D').mean().copy()
print(data_day.isna().sum())
plt.figure(figsize=(10,10))
sns.heatmap(data_day.corr(),annot=True)
```

```python
cols_mean = [f'{i}_mean' for i in data_day.columns]
data_day.columns = cols_mean
data_day
```

```python
corr_matrix = data_day.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
extract = [column for column in upper.columns if any(upper[column] > 0.7)]
print(extract.append('ActivePower_mean'))
```

```python
plt.figure(figsize=(10,50))
list_high_corr = ['','WindSpeed_mean']
data_day[extract].plot(subplots=True)
```

```python
data_update = data_day[extract]
if (data_update['Blade2PitchAngle_mean'].equals(data_update['Blade3PitchAngle_mean'])==True):
  data_update = data_update.drop('Blade3PitchAngle_mean', axis=1)
data_update = data_update.fillna(method='ffill').fillna(method='bfill')
data_update.isnull().sum()
```

```python
data_update.corr()
```

```python
((data_update-data_update.mean())/data_update.std()).iloc[:,-2:].plot()
```

```python
data_update_1 = data_update[['WindSpeed_mean','ActivePower_mean']]
data_update_1.index = data_update_1.index.date
data_update_1.to_csv('data_update_1.csv')
```
